```python
# Contar valores nulos por columna
# Crear un DataFrame con la información de nulos
nulos_info = pd.DataFrame({
    "columna":df.columns,
    "valores_nulos":df.isnull().sum(),
    "porcentaje": (df.isnull().mean() * 100)
}).sort_values(by="valores_nulos", ascending=False)

print(nulos_info.to_string(index=False))

# Resumen
print("\nResumen del dataset:")
total_nulos = df.isnull().sum().sum()
total_celdas = df.shape[0] * df.shape[1]
print(f"📌 Total de valores nulos en el dataset: {total_nulos:,}")
print(f"📌 Porcentaje total de nulos: {(total_nulos / total_celdas * 100):.2f}%")
```

```python
# Importar librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

# Configuraciones
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
plt.style.use('seaborn-v0_8-darkgrid')
```

```python
# Análisis detallado de nulos
print("❓ ANÁLISIS DE VALORES NULOS")
print("=" * 80)

nulos = df.isnull().sum()
nulos_pct = (nulos / len(df) * 100).round(2)

nulos_df = pd.DataFrame({
    'Columna': nulos.index,
    'Nulos': nulos.values,
    'Porcentaje': nulos_pct.values
}).query('Nulos > 0').sort_values('Nulos', ascending=False)

print(nulos_df.to_string(index=False))

# Visualizar filas con nulos
print("\n📋 Ejemplos de registros con valores nulos:")
print(df[df.isnull().any(axis=1)].head(10))
```

```python
#Tomamos decisiones sobre el manejo de valores nulos con base en el análisis anterior.

#1. Eliminamos la variable de suscriptores(suscribers) en los ultimos 30 días ya que tiene muchos valores nulos y para nuestro analisis no es relevante.
df_clean = df.drop(columns=['subscribers_for_last_30_days'])

#2. Empezamos con los porcentajes con 12% de nulos. Nos centramos en la variable de country, en este caso vamos a imputar los valores nulos con "Unknown" ya que no conocemos mas informacion.
df_clean['Country'] = df_clean['Country'].fillna('Unknown')
```

```python
#Revisamos como va el dataset 
df_clean.head()
```

```python
#3. Seguimos con la variable de abreviacion(abbreviation),vamos a imputar
df_clean['Abbreviation'] = df_clean['Abbreviation'].fillna('Unknown')
```

```python
#4.Debido a que sabemos que queremos analizar y que variables nos sirven, vamos a enfocarnos solo en esas. El resto las eliminamos, mas adelante.

#Seguimos con la variable de vistas de video de los ultimos 30 dias(video_views_for_last_30_days), en este caso vamos a imputar con la mediana de la columna.
df_clean['video_views_for_the_last_30_days'] = df_clean['video_views_for_the_last_30_days'].fillna(df_clean['video_views_for_the_last_30_days'].median())
```

```python
#5.Vamos a unir en una nueva columna con las variables de category y channel_type

for col in ["category", "channel_type"]:
    df_clean[col] = (
        df[col]
        .astype(str)
        .str.strip()
        .str.title()
        .replace("Nan", np.nan)
    )

#Creamos la nueva columna combinada
df_clean["content_type"] = (
    df["category"]
    .combine_first(df["channel_type"])
)

#Y si llegase a faltar algun valor nulo, lo imputamos con "Unknown"
df_clean["content_type"] = df_clean["content_type"].fillna("Unknown")
```

```python
#6. Vamos a crear una nueva variable uniendo las variables de created_date, created_year y created_month
#Empezaremos por convertir la variable created_date y created_year a formato datetime para que no salgan decimales
#Y convertimos created_month a string para que no de error al unirlas

df_clean["created_year"] = df["created_year"].astype("Int64")
df_clean["created_date"] = df["created_date"].astype("Int64")

month_map = {
    "January": 1, "Jan": 1,
    "February": 2, "Feb": 2,
    "March": 3, "Mar": 3,
    "April": 4, "Apr": 4,
    "May": 5,
    "June": 6, "Jun": 6,
    "July": 7, "Jul": 7,
    "August": 8, "Aug": 8,
    "September": 9, "Sep": 9,
    "October": 10, "Oct": 10,
    "November": 11, "Nov": 11,
    "December": 12, "Dec": 12
}

df_clean["created_month_num"] = (
    df_clean["created_month"]
    .astype(str)
    .str.strip()
    .str.title()
    .map(month_map)
    .astype("Int64")
)

```

```python
#Convertimos las columnas a tipo numerico para asegurarnos
df_clean["created_year"] = pd.to_numeric(df_clean["created_year"], errors="coerce")
df_clean["created_month_num"] = pd.to_numeric(df_clean["created_month_num"], errors="coerce")
df_clean["created_date"] = pd.to_numeric(df_clean["created_date"], errors="coerce")

#Creamos la nueva columna de fecha completa
df_clean["created_full_date"] = pd.to_datetime(
    df_clean["created_year"].astype("Int64").astype(str) + "-" +
    df_clean["created_month_num"].astype("Int64").astype(str) + "-" +
    df_clean["created_date"].fillna(1).astype("Int64").astype(str),
    errors="coerce"
)

```

```python
#7.Añadimos variables nuevas para nuestro analisis

#Creamos columnna de cuantos dias tiene el canal desde su creacion
df_clean["channel_age_days"] = (
    pd.Timestamp.today() - df_clean["created_full_date"]
).dt.days

df_clean["created_year"] = df_clean["created_full_date"].dt.year
df_clean["created_month"] = df_clean["created_full_date"].dt.month
df_clean["created_weekday"] = df_clean["created_full_date"].dt.weekday

df_clean["channel_age_days"] = df_clean["channel_age_days"].astype("Int64")
df_clean["created_weekday"] = df_clean["created_weekday"].astype("Int64")
```
